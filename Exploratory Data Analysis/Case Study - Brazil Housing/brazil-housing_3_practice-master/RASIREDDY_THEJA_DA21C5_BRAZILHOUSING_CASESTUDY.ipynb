```python
import pandas as pd

raw_data = pd.read_csv('DS1_C5_S3_BazilHousing_Data_Hackathon.csv')

raw_data
```

```python
# Remove duplicates
cleaned_data = raw_data.drop_duplicates().reset_index(drop=True)
```

```python
cleaned_data
```

```python
# Exclude Rio de Janeiro and Sao Paulo
filtered_data = cleaned_data[~cleaned_data['city'].isin(['Rio de Janeiro', 'São Paulo'])]
```

```python
filtered_data
```

```python
filtered_data.to_csv('filtered_housing_data.csv', index=False)

# Reload the filtered data
filtered_data = pd.read_csv('filtered_housing_data.csv')
```

```python
filtered_data
```

```python
# Analysis for bachelors
bachelor_cities = filtered_data[
    (filtered_data['rooms'] == 1) &
    (filtered_data['bathroom'] == 1) &
    (filtered_data['parking spaces'].between(0, 1)) &
    (filtered_data['furniture'] == 'not furnished') &
    (filtered_data['animal'] == 'acept') &
    (filtered_data['parking spaces'] == 1)
]

average_total_rent_bachelor = bachelor_cities['total (R$)'].mean()

print("Cities suitable for bachelors:")
print(bachelor_cities[['city', 'animal', 'parking spaces', 'total (R$)']])
print("Average total rent for bachelor cities:", average_total_rent_bachelor, "R$")
print("========================================")

# Analysis for mid-sized families
family_cities = filtered_data[
    (filtered_data['rooms'].between(2, 3)) &
    (filtered_data['bathroom'].between(1, 2)) &
    (filtered_data['parking spaces'].between(1, 2)) &
    (filtered_data['furniture'] == 'not furnished') &
    (filtered_data['animal'] == 'acept') &
    (filtered_data['parking spaces'] == 2)
]

average_total_rent_family = family_cities['total (R$)'].mean()
print("Cities suitable for mid-sized families:")
print(family_cities[['city', 'animal', 'parking spaces', 'total (R$)']])
print("Average total rent for family cities:", average_total_rent_family, "R$")
print("========================================")

# Analysis for large families
large_family_cities = filtered_data[
    (filtered_data['rooms'].between(4, 5)) &
    (filtered_data['bathroom'].between(2, 5)) &
    (filtered_data['parking spaces'].between(2, 5)) &
    (filtered_data['furniture'] == 'not furnished') &
    (filtered_data['animal'] == 'acept') &
    (filtered_data['parking spaces'] == 4)
]

average_total_rent_large_family = large_family_cities['total (R$)'].mean()
print("Cities suitable for large families:")
print(large_family_cities[['city', 'animal', 'parking spaces', 'total (R$)']])
print("Average total rent for large family cities:", average_total_rent_large_family, "R$")
```
