```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv = pd.read_excel('DS1_C5_S2_Income_Data_Challenge.xlsx')
csv
```

# Task 1

```python
cat_data = csv.select_dtypes(include='object')
num_data = csv.select_dtypes(exclude = 'object')
```

```python
cat_data
```

```python
num_data
```

# Task 2

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Marital_Status', hue='Income_Bracket', data=cat_data)
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Occupation', hue='Gender', data=cat_data)
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 15))
sns.set(style='darkgrid')
sns.catplot(x='Marital_Status',hue='Income_Bracket',data=cat_data, kind='count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Task 3

```python
plt.figure(figsize=(10, 8))
cor_mat = num_data.corr()
sns.heatmap(cor_mat, annot=True, fmt=".3f", linewidths=0.5)
plt.show()
```
