### Task - 1

```python
# Seprating out the categorical and continuous variables
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in data.columns:                # looping on the number of columns
        if data[column].dtype == 'object':       # nunique:  returns unique values in the column,
                                                    #here 100 is the hyperparamter and can be changed
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(data)         # Calling the function

# # Tabulate is a package used to print the list, dict or any data sets in a proper format; in table format
from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

### Task - 3

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for attrited employees traveling rarely
attrited_rare_travel = data[(data['Attrition'] == 'Yes') & (data['BusinessTravel'] == 'Travel_Rarely')]

# Count of attrited employees by job role
plt.figure(figsize=(10, 6))
sns.countplot(y = 'JobRole', data=attrited_rare_travel, order=attrited_rare_travel['JobRole'].value_counts().index)
plt.title('Count of Attrited Employees Traveling Rarely by Job Role')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.show()
```

### Task - 4

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for employees with education in Life Sciences and Medical background
life_medical_education = data[data['EducationField'].isin(['Life Sciences', 'Medical'])]

# Count of employee attrition by job role for Life Sciences and Medical background
plt.figure(figsize=(10, 6))
sns.countplot(y='JobRole', hue='Attrition', data=life_medical_education, order=life_medical_education['JobRole'].value_counts().index)
plt.title('Employee Attrition from Life Sciences and Medical Education Background by Job Role')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.legend(title='Attrition', loc='lower right')
plt.show()
```
