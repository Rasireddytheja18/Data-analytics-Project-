```python
# Define the available items and their quantities and costs
items = {
    "Biscuits": {"Quantity": 10, "Cost": 2},
    "Cereals": {"Quantity": 15, "Cost": 7},
    "Chicken": {"Quantity": 12, "Cost": 13},
    "Oats": {"Quantity": 3, "Cost": 3},
    "Rice": {"Quantity": 4, "Cost": 10}
}

# Function to calculate delivery charge based on distance
def calculate_delivery_charge(distance):
    if distance <= 5:
        return "$10"
    elif 5 < distance <= 10:
        return "$12"
    elif 10 < distance <= 15:
        return "$14"
    else:
        return "$20"

# Function to display the cost of purchased items
def display_cost_of_purchased_items(cart, items):
    total_cost = sum(cart[item] * items[item]['Cost'] for item, details in items.items() if item in cart)
    print(f"Total Cost: ${total_cost}")

# Main shopping loop
while True:
    print("\n--- Welcome to the Store! ---")
    print("Available items:")
    
    # Display available items and their details
    for item, details in items.items():
        print(f"{item}: {details['Quantity']} in stock, Cost: ${details['Cost']} each")

    # Prompt user for information
    customer_name = input("Enter your name: ")
    delivery_address = input("Enter your delivery address: ")
    distance = float(input("Enter the distance from the store in miles: "))
    
    cart = {}
    for item, details in items.items():
        quantity = int(input(f"How many {item}s would you like to buy? (0 for none): "))
        
        # Check if the requested quantity is available
        if quantity > details['Quantity']:
            print(f"Sorry, we only have {details['Quantity']} {item}s in stock.")
        else:
            cart[item] = quantity
            details['Quantity'] -= quantity

    # Display the cost of purchased items
    display_cost_of_purchased_items(cart, items)

    total_cost = sum(cart[item] * details['Cost'] for item, details in items.items() if item in cart)
    delivery_charge = calculate_delivery_charge(distance)

    print("\n--- Invoice ---")
    print(f"Customer Name: {customer_name}")
    print(f"Delivery Address: {delivery_address}")
    print(f"Delivery Charge: {delivery_charge}")
    print("Items Purchased:")
    for item, quantity in cart.items():
        print(f"{item}: {quantity} (Cost: ${items[item]['Cost']})")
    print(f"Total Cost: ${total_cost + (0 if delivery_charge == '$10' else 5)}\n")

    cont = input("Do you want to continue shopping? (yes/no): ")
    if cont.lower() != "yes":
        break
```
