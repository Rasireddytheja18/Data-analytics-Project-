```python
import mysql.connector
import pandas as pd
import pymysql
from sqlalchemy import create_engine


housing_data = pd.read_csv('DS1_C4_S6_House_Prices_Data_Challenge.csv')

housing_data
```

```python
filtered_data = housing_data[(housing_data['LotArea'] >= 6000) & (housing_data['LotConfig'] == 'Corner')]

filtered_data
```

```python
Task 1
```

```python
engine = create_engine('mysql+pymysql://root:root123@localhost/hr')
table_name = 'corner_plot_houses'
filtered_data.to_sql(table_name, con=engine, index=False, if_exists='append')
```

```python
# task 2
```

```python
filtered_data = housing_data[
    (housing_data['BldgType'] == '1Fam') &  # One family
    (housing_data['HouseStyle'].str.contains('1Story', case=False)) &  # One story
    (housing_data['LandContour'] == 'Lvl')  # Flat land, avoiding mountains or valleys/banks
]
table_name = 'one_family_one_story_flat_land'
filtered_data.to_sql(table_name, con=engine, index=False, if_exists='append')
```

```python
# task 3
```

```python
filtered_data = housing_data[housing_data['BldgType'].str.contains('Twnhs', case=False)]
filtered_data
```

```python
table_name = 'compact_townhouses'
filtered_data.to_sql(table_name, con=engine, index=False, if_exists='append')
```

```python
# task 4
```

```python
filtered_data = housing_data[
    (housing_data['BedroomAbvGr'] >= 3) & 
    (housing_data['BsmtQual'] == 'Ex') &  
    (housing_data['Exterior1st'] == 'CemntBd')
]
table_name = 'popular_houses'

filtered_data.to_sql(table_name, con=engine, index=False, if_exists='append')
```

```python
# task 5
```

```python
filtered_data = housing_data[
    (housing_data['GarageCars'] == 2) &  # Two cars in the garage
    (housing_data['GarageArea'] >= 500)  # Garage area is at least 500 square feet
]
table_name = 'two_car_garages'
filtered_data.to_sql(table_name, con=engine, index=False, if_exists='append')
```
