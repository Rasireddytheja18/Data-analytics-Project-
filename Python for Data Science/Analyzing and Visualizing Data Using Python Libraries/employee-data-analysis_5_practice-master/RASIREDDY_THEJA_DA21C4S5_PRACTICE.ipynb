```python
import pandas as pd

data = pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
data
```

### Task - 1

```python
emp_sal = data.groupby(by = ['Department'])['Annual Salary ($)'].sum()
emp_sal
```

```python
import matplotlib.pyplot as plt

emp_sal.plot(kind='bar', color='skyblue')
                                           # To build bar chart
plt.title("Vertical Bar chart")            # Title name
plt.show()
```
