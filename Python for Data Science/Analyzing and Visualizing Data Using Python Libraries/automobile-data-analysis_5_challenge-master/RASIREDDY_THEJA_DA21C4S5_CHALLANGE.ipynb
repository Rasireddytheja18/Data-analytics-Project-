```python
import pandas as pd

data = pd.read_excel('DS1_C4_S5_Car_Data_Challenge.xlsx')
data
```

### Task - 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by manufacturer, calculate mean city mileage, and sort the values
manufacturer_mileage = data.groupby(by = 'Make')['City_Mileage_km_litre'].mean().sort_values(ascending=False)

# Get the top 10 manufacturers
top_10_manufacturers = manufacturer_mileage.head(10)

# Plotting the horizontal bar graph

top_10_manufacturers.plot(kind='bar', color='skyblue')
plt.title('Top 10 Car Manufacturers based on City Mileage')
plt.xlabel('City Mileage')
plt.ylabel('Manufacturer')
plt.show()
```

### Task - 5

```python
# Filter data for the specified manufacturers
filtered_data = data[(data.Make == 'Hyundai') | (data.Make == 'Mahindra') | (data.Make == 'Renault') | (data.Make == 'Skoda')]


filtered_data
```

```python
mean_city_mileage = filtered_data.groupby('Make')['City_Mileage_km_litre'].mean()
mean_city_mileage
```

```python
median_city_mileage = filtered_data.groupby('Make')['City_Mileage_km_litre'].median()
median_city_mileage
```

```python
mileage_summary = pd.DataFrame({'Mean': mean_city_mileage, 'Median': median_city_mileage})
mileage_summary
```

```python
mileage_summary.plot(kind='bar', stacked=True, title='Stacked Bar Graph by dataframe')
plt.xlabel('Manufacturers')
plt.ylabel('Mileage')

plt.show()
```

### Task - 6

```python
data.isnull().sum().head(20)
```

```python
data['City_Mileage_km_litre'] = data['City_Mileage_km_litre'].fillna(data['City_Mileage_km_litre'].mean())
```

```python
data['Highway_Mileage_km_litre'] = data['Highway_Mileage_km_litre'].fillna(data['Highway_Mileage_km_litre'].mean())
```

```python
data.isnull().sum().head(20)
```

```python
avg_mileage = data.groupby('Make')[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']].mean()
avg_mileage
```

```python
avg_mileage['Total_Mileage'] = avg_mileage['City_Mileage_km_litre'] + avg_mileage['Highway_Mileage_km_litre']
avg_mileage['Total_Mileage']
```

```python
avg_mileage_sorted = avg_mileage.sort_values('Total_Mileage')
avg_mileage_sorted
```

```python
plt.figure(figsize=(10, 8))
plt.barh(avg_mileage_sorted.index, avg_mileage_sorted['Total_Mileage'], color='skyblue')
plt.title('Average Mileage of Manufacturers (City + Highway)')
plt.show()
```
