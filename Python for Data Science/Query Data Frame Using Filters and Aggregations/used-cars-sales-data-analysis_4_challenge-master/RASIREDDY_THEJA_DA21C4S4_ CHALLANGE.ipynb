```python
import pandas as pd
data = pd.read_csv('DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')
```

```python
data
```

```python
# task 1
```

```python
missing_values = data.isnull().sum()
data_cleaned = data.dropna()
print(missing_values,data_cleaned)
```

```python
# task 2
```

```python

potential_cars = data[(data['year'] == 2014) & (data['state'] == 'ca')]

potential_cars_sorted = potential_cars.sort_values(by='mp', ascending=True)

top_100_lowest_priced_cars = potential_cars_sorted.head(100)

print(top_100_lowest_priced_cars)
```

```python
# task 3
```

```python
cross_table = pd.crosstab(top_100_lowest_priced_cars['make'], 
                          top_100_lowest_priced_cars['mp'], 
                          values=top_100_lowest_priced_cars['model'], 
                          aggfunc='count', 
                          margins=True, 
                          margins_name='Total')
print(cross_table)
```

```python
# task 4
```

```python

chevrolet_cars = top_100_lowest_priced_cars[top_100_lowest_priced_cars['make'] == 'Chevrolet'].copy()

chevrolet_cars['Discounted_Price'] = chevrolet_cars['mp'] * 0.9

print(chevrolet_cars[['make', 'model', 'mp', 'Discounted_Price']])
```

```python
# task 5
```

```python

average_price_by_interior_color = top_100_lowest_priced_cars.groupby('interior')['mp'].mean()

print(average_price_by_interior_color)
```

```python
# task 6
```

```python

filtered_cars = data[(data['year'] >= 2010) & (data['year'] <= 2015)]

color_availability = filtered_cars.groupby(['make', 'color']).size().reset_index(name='Count')

print(color_availability)
```

```python
# task 7
```

```python
# Filter the data for BMW Sedans manufactured between 2010 and 2015 with condition score 5
bmw_sedans = data[(data['make'] == 'BMW') & (data['body'] == 'sedan') & (data['year'] >= 2010) & (data['year'] <= 2015) & (data['condition'] == 5)]

# Display the details of the BMW Sedans
print(bmw_sedans)

# Group by Year and Model, then find the maximum expected price (mp)
year_wise_max_price = bmw_sedans.groupby(['year', 'model'])['mp'].max().reset_index()

# Display the year-wise maximum expected price for each model
print(year_wise_max_price)
```

```python
# task 8
```

```python

suvs = data[(data['body'] == 'suv') & (data['year'] >= 2010) & (data['year'] <= 2015) & (data['odometer'] <= 30000)]

top_100_suvs = suvs.sort_values(by='mp', ascending=False).head(100)

print(top_100_suvs)
```
